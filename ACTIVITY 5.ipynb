#1A Trapezoid Rule

```python
import numpy as np
a = 1
b = 2
N = 349320
h = (b-a)/(N-1)
def func(x):
    return np.cos(3*x)
sum = (func(a) + func(b))/2
for i in range (1, N-1):
    sum = sum + func(a + i*h)
area = sum*h
print("Trapezoid rule =", area)
print("Exact value    =", (np.sin(3*2)- np.sin(3*1))/3)
```

#1B Simpson Rule

```python
import numpy as np 
a = 1
b = 2 
N = 100000
h = (b-a)/(N-1)
def func(x):
    return np.cos(3*x)
term1 = 0
term2 = 0

sum = (func(a) + func(b))/3
for i in range (0, N-1, 2):
    fx = a + i*h
    if i % 2 == 0:
        term1 = term1 + fx
    else:
        term2 = term2 + fx
area = h*(sum + 2*term1 + 4*term2)/3
print ("Simpson rule =", area)
print ("Exact value  =", (np.sin(3*2) - np.sin(3*1))/3)
```

Plotting a function

```python
import matplotlib.pyplot as plt
x = np.linspace (0, 10, 100)
fx = np. sin (x)
plt.plot (x, fx)
gx = np.cos(x)
plt.plot (x, gx)
h = 0.1
deriv = ((np.sin(x+(h/2)) - (np.sin(x-(h/2)))) / h)
plt.scatter(x, deriv)
plt.show()
```
