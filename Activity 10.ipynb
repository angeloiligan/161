Harmonic Oscillation

```python
import numpy as np 
from numpy.linalg import eig 
import matplotlib.pyplot as plt

hbar = 1.054571817*10**-34
m = 9.1093837*10**-31
N = 10 #number of points 
omega = 1.0
x = np.linspace(0, 10, N)
dx = x[1] - x[0] #grid spacing
t0 = -hbar**2/2*m*dx**2
H = np.zeros((N, N))

def v(x): #potential 
    return (1/2)*m*omega**2*x**2
for i in range (0, N):
    H[i, i] = 2*t0 + v(x[i])
for i in range (0, N-1):
    H[i, i+1] = -t0
for i in range (1, N-1):
    H[i, i-1] = -t0
print (H)

v, w = eig(H)
d = np.argsort(v)
v = v[d]
w = w[:, d]
print (v)

plt.figure(figsize=(10, 6))
plt.plot (x, w[0]**2, 'r-')
plt.plot (x, w[1]**2, 'g-')
plt.plot (x, w[2]**2, 'y-')
plt.plot (x, w[3]**2, 'b-')
plt.show()
```
