```python
import numpy as np 
import matplotlib.pyplot as plt 
def func(t, x):
    return -x

t0 = 0 
v0 = 0
x0 = 1
h = 0.1
t = np.zeros(100)
x = np.zeros(100)
v = np.zeros(100)
t[0] = t0
x[0] = x0
v[0] = v0

for i in range (0, 99):
    x[i+1] = x[i] + h*v[i]
    v[i+1] = v[i] + h*func(t[i], x[i])
    t[i+1] = t[i] + h

plt.plot (t, x, 'ro', label = 'Euler')
plt.plot (t, np.cos(t), label = 'Exact')
plt.legend()
plt.show()
```

Runge-Kutta Method 

```python
def func(t, x):
    return -x
    return (t, x)
t0 = 0 
v0 = 0
x0 = 1
h = 0.1
t = np.zeros(100)
x = np.zeros(100)
v = np.zeros(100)
t[0] = t0
x[0] = x0
v[0] = v0

for i in range (0, 99):
    k1 = h*func(t[i], x[i])
    k2 = h*func(t[i] + h/2, x[i] + k1/2)
    k3 = h*func(t[i] + h/2, x[i] + k2/2)
    k4 = h*func(t[i] + h, x[i] + k3)
    x[i+1] = x[i] + h*v[i]
    v[i+1] = v[i] + (1/6)*(k1 + 2*k2 + 2*k3 + k4)
    t[i+1] = t[i] + h

plt.plot (t, x, 'ro', label = 'Euler')
plt.plot (t, np.cos(t), label = 'Exact')
plt.legend()
plt.show()
```
