```python
import numpy as np
import matplotlib.pyplot as plt

def func(t, x, vx):
    return -k*vx/m  # with air resistance

def func2(t, y, vy):
    return -g  

theta = 55 * np.pi / 180  
m = 0.2                   
k = 0.08                  
g = 9.8                   

v0 = 1450
vx0 = v0 * np.cos(theta)
vy0 = v0 * np.sin(theta)
x0 = 0
y0 = 0

t = np.linspace(0, 7, 70000) 
h = t[1] - t[0]               
vx = np.zeros_like(t)
vy = np.zeros_like(t)
x = np.zeros_like(t)
y = np.zeros_like(t)

# Set initial conditions
vx[0] = vx0
vy[0] = vy0
x[0] = x0
y[0] = y0

for i in range(len(t) - 1):
    k1_vx = h * func(t[i], x[i], vx[i])
    k1_vy = h * func2(t[i], y[i], vy[i])

    k2_vx = h * func(t[i] + h/2, x[i] + vx[i] * h/2, vx[i] + k1_vx/2)
    k2_vy = h * func2(t[i] + h/2, y[i] + vy[i] * h/2, vy[i] + k1_vy/2)

    k3_vx = h * func(t[i] + h/2, x[i] + vx[i] * h/2, vx[i] + k2_vx/2)
    k3_vy = h * func2(t[i] + h/2, y[i] + vy[i] * h/2, vy[i] + k2_vy/2)

    k4_vx = h * func(t[i] + h, x[i] + vx[i] * h, vx[i] + k3_vx)
    k4_vy = h * func2(t[i] + h, y[i] + vy[i] * h, vy[i] + k3_vy)

    vx[i+1] = vx[i] + (1/6) * (k1_vx + 2*k2_vx + 2*k3_vx + k4_vx)
    vy[i+1] = vy[i] + (1/6) * (k1_vy + 2*k2_vy + 2*k3_vy + k4_vy)

    x[i+1] = x[i] + vx[i] * h
    y[i+1] = y[i] + vy[i] * h

plt.plot(x, y, 'y-', label='Projectile Trajectory')
plt.legend()
plt.show()
```
